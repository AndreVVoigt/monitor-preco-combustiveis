# Ingestão na Bronze

### Problemas dos Dados
- São arquivos .xlsx (Databricks tem problemas em ler)
- Nome das tabelas precisando serem normalizadas (espaço entre os nomes e caracteres nao convencionais como o '~')
- Algumas colunas não tinham um unico tipo de dado, para isso passei tudo para String e na Silver Layer tudo será corrigido

### Arquivos Mensais

```python
def sanitize_column(col):

    col = re.sub(r'[ ,;{}()\n\t=]', '_', col)
    col = col.encode('ascii', 'ignore').decode()
    col = col.lower()
    return col

path = "/Volumes/projeto_combustiveis/bronze/raw/mensal-brasil-desde-jan2013.xlsx"

df_pandas = pd.read_excel(path, skiprows=16)
df_pandas = df_pandas.astype(str)


new_columns = [sanitize_column(col) for col in df_pandas.columns]
df_pandas.columns = new_columns

df_spark = spark.createDataFrame(df_pandas)

df_spark.write.mode("overwrite").saveAsTable("projeto_combustiveis.bronze.mensal_brasil_desde_jan2013")
```

### Arquivo: resumo_semanal_lpc_2026-02-22_2026-02-28

- O arquivo possui múltiplas abas no Excel. Para facilitar o processamento, implementei um loop que transforma cada aba em uma tabela separada.

```python
path_semanal = "/Volumes/projeto_combustiveis/bronze/raw/resumo_semanal_lpc_2026-02-22_2026-02-28.xlsx"

abas_para_ingerir = ["BRASIL", "REGIOES", "ESTADOS", "MUNICIPIOS", "CAPITAIS"]

for aba in abas_para_ingerir:
    print(f"Processando a aba: {aba}...")
    
    df_pandas = pd.read_excel(path_semanal, sheet_name=aba, skiprows=9)
    df_pandas = df_pandas.astype(str)
    
    df_pandas.columns = [sanitize_column(col) for col in df_pandas.columns]
    
    df_spark = spark.createDataFrame(df_pandas)
    
    nome_tabela = sanitize_column(aba)
    
    df_spark.write.mode("overwrite").saveAsTable(f"projeto_combustiveis.bronze.semanal_{nome_tabela}")

print("Todas as tabelas semanais foram salvas na Bronze com sucesso!")
```
