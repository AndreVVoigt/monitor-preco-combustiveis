# Criando Fact historico


## Brasil

```python
df_fato_historico_brasil = spark.sql("""
    SELECT 
        t.id_tempo,
        p.id_produto,
        s.unidade_de_medida,
        s.numero_postos_pesquisados,
        s.preco_medio_revenda,
        s.desvio_padrao_revenda,
        s.preco_minimo_revenda,
        s.preco_maximo_revenda,
        s.margem_media_revenda,
        s.coef_variacao_revenda,
        s.preco_medio_distribuicao,
        s.desvio_padrao_distribuicao,
        s.preco_minimo_distribuicao,
        s.preco_maximo_distribuicao,
        s.coef_variacao_distribuicao
    FROM projeto_combustiveis.silver.historico_mensal_brasil s
    
    INNER JOIN projeto_combustiveis.gold.dim_tempo t ON s.mes_referencia = t.data_referencia
    
    INNER JOIN projeto_combustiveis.gold.dim_produto p ON s.produto = p.produto
""")
```

```python
df_fato_historico_brasil.write.mode("overwrite").saveAsTable("projeto_combustiveis.gold.fact_historico_brasil")
display(df_fato_historico_brasil)
```

## Regiões


```python
df_fato_regioes = spark.sql("""
    SELECT 
        t.id_tempo,
        l.id_localidade,
        p.id_produto,
        s.unidade_de_medida,
        s.numero_postos_pesquisados,
        s.preco_medio_revenda,
        s.desvio_padrao_revenda,
        s.preco_minimo_revenda,
        s.preco_maximo_revenda,
        s.margem_media_revenda,
        s.coef_variacao_revenda,
        s.preco_medio_distribuicao,
        s.desvio_padrao_distribuicao,
        s.preco_minimo_distribuicao,
        s.preco_maximo_distribuicao,
        s.coef_variacao_distribuicao
    FROM projeto_combustiveis.silver.historico_mensal_regioes s

    INNER JOIN projeto_combustiveis.gold.dim_tempo t ON s.mes_referencia = t.data_referencia

    INNER JOIN projeto_combustiveis.gold.dim_produto p ON s.produto = p.produto

    INNER JOIN projeto_combustiveis.gold.dim_localidade l 
        ON s.regiao = l.regiao 
        AND l.estado IS NULL 
        AND l.municipio IS NULL
""")
```

```python
df_fato_regioes.write.mode("overwrite").saveAsTable("projeto_combustiveis.gold.fact_historico_regioes")
display(df_fato_regioes)
```
