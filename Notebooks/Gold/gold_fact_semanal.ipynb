# Criando Fact Semanal


## Brasil

```python
df_fato_semanal_brasil = spark.sql("""
    SELECT 
        t.id_tempo, 
        p.id_produto, 
        s.unidade_de_medida, 
        s.numero_postos_pesquisados,
        s.preco_medio_revenda, 
        s.desvio_padrao_revenda, 
        s.preco_minimo_revenda, 
        s.preco_maximo_revenda,
        s.coef_variacao_revenda
    FROM projeto_combustiveis.silver.monitor_semanal_brasil s

    INNER JOIN projeto_combustiveis.gold.dim_tempo t ON s.data_referencia = t.data_referencia

    INNER JOIN projeto_combustiveis.gold.dim_produto p ON s.produto = p.produto
""")
```

```python
df_fato_semanal_brasil.write.mode("overwrite").saveAsTable("projeto_combustiveis.gold.fact_semanal_brasil")
display(df_fato_semanal_brasil)
```

## Regiões

```python
df_fato_semanal_regioes = spark.sql("""
    SELECT 
        t.id_tempo, 
        l.id_localidade, 
        p.id_produto, 
        s.unidade_de_medida, 
        s.numero_postos_pesquisados,
        s.preco_medio_revenda, 
        s.desvio_padrao_revenda, 
        s.preco_minimo_revenda, 
        s.preco_maximo_revenda,
        s.coef_variacao_revenda
    FROM projeto_combustiveis.silver.monitor_semanal_regioes s

    INNER JOIN projeto_combustiveis.gold.dim_tempo t ON s.data_referencia = t.data_referencia

    INNER JOIN projeto_combustiveis.gold.dim_produto p ON s.produto = p.produto

    INNER JOIN projeto_combustiveis.gold.dim_localidade l 
        ON s.regiao = l.regiao 
        AND l.estado IS NULL 
        AND l.municipio IS NULL
""")
```

```python
df_fato_semanal_regioes.write.mode("overwrite").saveAsTable("projeto_combustiveis.gold.fact_monitor_semanal_regioes")
display(df_fato_semanal_regioes)
```
